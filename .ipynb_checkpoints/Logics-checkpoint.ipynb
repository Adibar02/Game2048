```python
import random
def start_game():
    mat = [[0]*4 for i in range(4)]
    return mat

def add_new_2(mat):   
    row = random.randint(0,3)
    col = random.randint(0,3)
    while mat[row][col] != 0:
        row = random.randint(0,3)
        col = random.randint(0,3)
    mat[row][col] = 2
    return mat

def get_current_state(mat):
    won = False
    lose = True
    
    # if matrix has 2048 then won  
    for row in range(4):
        for col in range(4):
            if mat[row][col] == 0:
                lose = False
            if mat[row][col] == 2048:
                return 'WON'
            
    # if matrix is not full 
    for row in range(4):
        for col in range(4):
            if mat[row][col] == 0:
                return 'GAME NOT OVER'
    
    # if matrix contains something that can be added
        # every row n column except last
    for row in range(3):
        for col in range(3):
            if mat[row][col] == mat[i+1][j] or mat[row][col] == mat[i][j+1]:
                return 'GAME NOT OVER'
        # last row
    for col in range(3):
        if mat[3][col] == mat[3][col+1]:
            return 'GAME NOT OVER'
        # last col
    for row in range(3):
        if mat[row][3] == mat[row+1][3]:
            return 'GAME NOT OVER'
        
    return 'LOST'

def compress(mat):
    new_mat = []
    for i in range(4):
        new_mat.append([0]*4)
    for i in range(4):
        pos = 0
        for j in range(4):
            if mat[i][j] != 0:
                new_mat[i][pos] = mat[i][j]
                pos += 1
    return new_mat

def merge(mat):
    for row in range(4):
        for col in range(3):
            if mat[row][col] == mat[row][col+1] and mat[row][col] != 0:
                mat[row][col] = mat[row][col]*2
                mat[row][col+1] = 0
            
def reverse(mat):
    new_mat = []
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[i][3-j])

    return new_mat

def transpose(mat):
    new_mat = []
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[j][i])

    return new_mat


def move_up(mat):
    mat = transpose(mat)
    mat,changed1 = compress(mat)
    mat,changed2 = merge(mat)
    mat,changed3 = compress(mat)
    changed = changed1 or changed2
    mat = transpose(mat)
    return mat,changed
    
def move_down(mat):
    mat = transpose(mat)
    mat = reverse(mat)
    mat,changed1 = compress(mat)
    mat,changed2 = merge(mat)
    mat,changed3 = compress(mat)
    changed = changed1 or changed2
    mat = reverse(mat)
    mat = transpose(mat)
    return mat,changed

def move_left(mat):
    mat,changed1 = compress(mat)
    mat,changed2 = merge(mat)
    mat,changed3 = compress(mat)
    changed = changed1 or changed2
    return mat,changed
    
def move_right(mat):
    mat = reverse(mat)
    mat,changed1 = compress(mat)
    mat,changed2 = merge(mat)
    mat,changed3 = compress(mat)
    changed = changed1 or changed2
    mat = reverse(mat)
    return mat,changed
```
